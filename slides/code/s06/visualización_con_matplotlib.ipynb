```python
plt.style.use('fivethirtyeight')
N = 5
np.random.seed(N)

fig, (ax1, ax2) = plt.subplots(2, figsize=(10, 15))
men_means = np.random.randint(170, 195, N)
women_means = np.random.randint(150, 180, N)

men_std = np.random.uniform(0, 4, N)
women_std = np.random.uniform(0, 4, N)


ind = np.arange(N)

width = (men_std + women_std)/2/N

#####
ax1.bar(ind, men_std, width, color='#0055ff')
ax1.bar(ind, women_std, width, color='#fabada',
          bottom=men_std)
#####
ax1.set_ylabel('Scores')
ax1.set_xlabel('Groups')
ax1.set_title('Height std by group and gender')

ax1.set_xticklabels(('G1', 'G2', 'G3', 'G4', 'G5'))
ax1.legend(['men', 'women'], loc='best')

ind = ['G1', 'G2', 'G3', 'G4', 'G5']
######
ax2.barh(ind, men_means, width, color='#0055ff')
ax2.barh(ind, women_means, width, color='#fabada')
######
ax2.legend(['men', 'women'], loc='upper left')
ax2.set_title('Height mean by group and gender')
ax2.set_ylabel('Groups');
```

```python
from matplotlib import colors
from sklearn.datasets import load_wine

wine = load_wine()
vino = [wine.feature_names.index(param) for param in ['alcohol', 'flavanoids', 'hue', 'color_intensity']]
colores = np.stack(( wine.data[:, vino[2]]/10, len(wine.data)*[.99], len(wine.data)*[.5] ))  # hsv
tamao = wine.data[:, -1]*2
fig, ax = plt.subplots()

ax.scatter(wine.data[:,vino[0]],
           wine.data[:,vino[1]],
#            c=wine.target,
           color=list(map(colors.hsv_to_rgb, colores.T)),
           alpha=.7,
           s=tamao)

fig.suptitle('Niveles de alcohol y flavonoides (muestra de 178 vinos)')
ax.set_title('Vinos de 3 clases')
ax.set_xlabel('Alcohol ()')
ax.set_ylabel('Flavonoides (ppm)')

ax.annotate('Via Arcada', xy=(11.8, 1), xytext=(11, .5),
            arrowprops=dict(facecolor='black', shrink=0.05));
```
