```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
# Estilos de los gráficos

plt.style.use('ggplot')
# Para ver todos los estilos disponibles:
# print(plt.style.available)

plt.rcParams['figure.figsize'] = [15, 5]
```

```python
# Obtención de datos
N = 365
np.random.seed(2983)  # reproducibilidad
t = pd.date_range('1/1/2018', periods=N)
y = 200 * (np.random.randn(N).cumsum() + 40)

#####
fig = plt.figure()  # crea figura
ax = plt.subplot(1, 1, 1)  # crea ejes

ax.plot(t, y);  # dibuja sobre los ejes
#####
```

```python
#####
(fig, ax) = plt.subplots(1, 1)  # crea figura y ejes en 1 sólo paso

ax.plot(t, y);
#####
```

## `plt.subplots()`

```python
z = abs(8000 - y[::-1])/100

#####
# Crear figura y ejes (1 fila y 2 columnas)
fig, (eje1, eje2) = plt.subplots(1, 2)

eje1.plot(t, y,  # ejes (x, y)
          'g')   # tipo de línea(green)  <-- opcional
eje2.plot(t, z)
#####

eje1.set_title('Cotización criptomoneda en $')
eje2.set_title('Rentabilidad a futuro')

fig, (eje1, eje2) = plt.subplots(
    2, 1,  # 2 filas por 1 columna
    figsize=(8, 10),  # 8x10" de tamaño total
    sharex=True,  # mismo eje 'x'
    facecolor='silver')  # color de fondo

fig.suptitle('2 gráficos en vertical')

eje1.plot(t, y,  # (x, y)
          'g.')  # punteado en verde

eje2.plot(t, z,
          '.-');  # punto-línea
```

## `plt.subplot()`

```python
fig = plt.figure(figsize=(7, 7))

#####
ax1 = plt.subplot(2, 1, 1)  # 2 filas, 1 columna, seleccionar la posición 1
#####
ax1.plot(t, y,
         'gs')  # verde con marcas cuadradas (green, square)

#####
ax2 = plt.subplot(2, 2, 3)  # 2 filas, 2 columas, seleccionar la posición 3
#####
ax2.hist(y,
         alpha=.6)  # transparente al 60%
#####
ax3 = plt.subplot(2, 2, 4)  # 2 filas, 2 columas, seleccionar la posición 4
#####
ax3.pcolormesh(np.tril(np.random.uniform(size=(10, 10)), -1));
```
